### Question 4

```python
# Create new variable: "max_order"
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check new variable
ords_prods_merge.head(100)
```

```python
# Create loyalty flag
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check new flag
ords_prods_merge['loyalty_flag'].value_counts()
```

```python
# Check new df
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

### Question 5

```python
# Check basic statistics of loyalty flag variable and prices: mean, min, max
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

#### There is a slight difference in the spending habits of the 3 types of customers if you look at the average price of the items that they order. I don't think the difference is statistically significant though. It seems the new customers or those with the fewest number of orders buy more expensive items on average and the loyal customers buy cheaper items on average. It does seem however that all 3 types of customers have bought the cheapest item and the most expensive items in the shop. I would say that the difference is not significant when looking at the prices of items they buy.

### Question 6

```python
# Create new variable: avg_spending
ords_prods_merge['avg_spending'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(10)
```

```python
# Create spending flag
ords_prods_merge.loc[ords_prods_merge['avg_spending'] < 10, 'spending_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['avg_spending'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check new flag
ords_prods_merge['spending_flag'].value_counts()
```

### Question 7

```python
# Create new variable: order_frequency
ords_prods_merge['order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head(10)
```

```python
# Create frequency flag
ords_prods_merge.loc[ords_prods_merge['order_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['order_frequency'] <= 20) & (ords_prods_merge['order_frequency'] > 10), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['order_frequency'] <= 10, 'frequency_flag'] = 'frequent customer'
```

```python
# Check new flag
ords_prods_merge['frequency_flag'].value_counts()
```
